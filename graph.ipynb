```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('Assignment4_SalaryData.csv')
#print(df)
```

```python
ax = df.plot.area(title='Experience V/s Salary Comparesion')
ax.set(xlabel='Years of Experience', ylabel='Salary')
```

```python
ax=df.plot.line(title='Experience V/s Salary Comparesion')
ax.set(xlabel='Years of Experience', ylabel='Salary')
```
